```python
import pandas as pd
import pickle
import plotly.express as px

# Load the dataset
df = pd.read_csv('SecondCar.csv')

# Load the model pipeline from file
pipeline = pickle.load(open('Final_Model_Pipeline.sav', 'rb'))

def get_user_input():
    # Get user input for car name
    car_name = input("Enter Car Name (Available options: {}): ".format(', '.join(df['name'].unique())))
    
    if car_name not in df['name'].unique():
        print("Car name not found in the dataset.")
        return None

    # Filter the dataset based on the selected car name
    car_data = df[df['name'] == car_name]

    # Get unique options for the selected car
    fuel_types = car_data['fuel'].unique()
    seller_types = car_data['seller_type'].unique()
    transmission_types = car_data['transmission'].unique()
    owner_types = car_data['owner'].unique()
    years = car_data['year'].unique()
    ex_showroom_prices = car_data['ExShowroom Price'].unique()

    # Input the car features
    fuel_type = input("Enter Fuel Type (Available options: {}): ".format(', '.join(fuel_types)))
    seller_type = input("Enter Seller Type (Available options: {}): ".format(', '.join(seller_types)))
    transmission_type = input("Enter Transmission Type (Available options: {}): ".format(', '.join(transmission_types)))
    owner_type = input("Enter Owner Type (Available options: {}): ".format(', '.join(owner_types)))
    year = input("Enter Year (Available options: {}): ".format(', '.join(map(str, years))))

    try:
        km_driven = float(input("Enter Kilometers Driven: "))
        exshowroom_price = float(input("Enter ExShowroom Price (Available options: {}): ".format(', '.join(map(str, ex_showroom_prices)))))
    except ValueError:
        print("Please enter valid numbers for kilometers driven and ExShowroom Price.")
        return None

    return {
        'name': car_name,
        'year': year,
        'km_driven': km_driven,
        'fuel': fuel_type,
        'seller_type': seller_type,
        'transmission': transmission_type,
        'owner': owner_type,
        'ExShowroom Price': exshowroom_price
    }

def predict_price(input_data):
    # Create DataFrame for prediction
    input_df = pd.DataFrame([input_data])
    
    # Predict selling price
    predicted_price = pipeline.predict(input_df)
    return predicted_price[0]

def visualize_data(input_data):
    # Visualizations
    print("\nGenerating Visualizations...\n")

    # 2D Scatter Plot: ExShowroom Price vs Selling Price
    fig_scatter = px.scatter(df, x='ExShowroom Price', y='selling_price', color='fuel',
                             title="ExShowroom Price vs Selling Price",
                             labels={'ExShowroom Price': 'ExShowroom Price', 'selling_price': 'Selling Price'})
    fig_scatter.show()

    # 2D Line Plot: Year vs Selling Price
    fig_line = px.line(df, x='year', y='selling_price', color='seller_type',
                       title="Year vs Selling Price",
                       labels={'year': 'Year', 'selling_price': 'Selling Price'})
    fig_line.show()

    # Bar Plot: Fuel Type vs Average Selling Price
    fuel_avg_price = df.groupby('fuel')['selling_price'].mean().reset_index()
    fig_bar = px.bar(fuel_avg_price, x='fuel', y='selling_price',
                     title="Fuel Type vs Average Selling Price",
                     labels={'fuel': 'Fuel Type', 'selling_price': 'Average Selling Price'})
    fig_bar.show()

    # Predicted vs Actual Selling Prices for the selected car
    car_sales_data = df[df['name'] == input_data['name']]
    car_sales_data['Predicted Price'] = pipeline.predict(car_sales_data)

    # 2D Scatter Plot: Actual vs Predicted Selling Price
    fig_pred_actual = px.scatter(car_sales_data, x='selling_price', y='Predicted Price', color='year',
                                  title="Actual vs Predicted Selling Price",
                                  labels={'selling_price': 'Actual Selling Price', 'Predicted Price': 'Predicted Selling Price'})
    fig_pred_actual.show()

def main():
    # Main function to execute the program
    user_input = get_user_input()
    if user_input is not None:
        predicted_price = predict_price(user_input)
        print(f"\nPredicted Selling Price: â‚¹{round(predicted_price, 2)}")
        visualize_data(user_input)

if __name__ == "__main__":
    main()
```
